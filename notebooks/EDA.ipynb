```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.under_sampling import RandomUnderSampler
```

```python
file = '../data/raw/creditcard.csv'
data = pd.read_csv(file)
```

```python
sns.countplot(x=data['Class'])
plt.title('Distribuição das Classes')
plt.show()
```

Summary:

The dataset doesn't have any missing values.

The transaction time and amounts vary widely, with amounts being highly skewed as indicated by the histogram and box plot.

The box plot shows that fraud transactions (Class 1) tend to have different transaction amount distributions compared to normal transactions.

```python
plt.figure(figsize=(10,5))
sns.histplot(data['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amount')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(12, 8))
mask = np.triu(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            mask=mask, linewidths=0.5, annot_kws={"size": 7})
plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.tight_layout()
plt.show()
```

Overall, the EDA has shown that the dataset has no missing data, is highly imbalanced in terms of the target variable, and contains a mix of transformed features with various statistical distributions.

```python
x = data.drop(columns=['Class'])
y = data['Class']

undersampler = RandomUnderSampler(sampling_strategy=1.0, random_state=42)
x_resampled, y_resampled = undersampler.fit_resample(x, y)
```

```python
df_resampled = pd.DataFrame(x_resampled, columns=x.columns)
df_resampled['Class'] = y_resampled
```

```python
df_resampled.to_csv('../data/processed/creditcard_balanced.csv', index=False)
```
