Summary:

The dataset doesn't have any missing values.

The transaction time and amounts vary widely, with amounts being highly skewed as indicated by the histogram and box plot.

The box plot shows that fraud transactions (Class 1) tend to have different transaction amount distributions compared to normal transactions.

```python
# Visualizations: distribution of Amount feature
plt.figure(figsize=(10,5))
sns.histplot(data['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amount')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize correlation heatmap of the first 10 features (excluding Time and Class for readability)
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap for first 10 features')
plt.show()
```

```python
from imblearn.under_sampling import RandomUnderSampler

X = data.drop(columns=['Class'])
y = data['Class']

undersampler = RandomUnderSampler(sampling_strategy=1.0, random_state=42)
X_resampled, y_resampled = undersampler.fit_resample(X, y)
```

```python
# Verificando a distribuiÃ§Ã£o apÃ³s o undersampling
print("DistribuiÃ§Ã£o apÃ³s o undersampling:")
print(y_resampled.value_counts())
```

```python
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['Class'] = y_resampled
```

```python
# Ver nova distribuiÃ§Ã£o
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df_resampled, palette='viridis')
plt.title('DistribuiÃ§Ã£o das Classes ApÃ³s Undersampling')
plt.show()
```

```python
df_resampled.to_csv('../data/processed/creditcard_balanced.csv', index=False)
```
