```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
file = '../data/processed/creditcard_balanced.csv'
df = pd.read_csv(file)
```

```python
print('Head of the dataset:')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Distribution of Class variable
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df, palette='viridis')
plt.title('Distribution of Class (0: non-fraud, 1: fraud)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Amount feature by Class
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df[df['Class']==0]['Amount'], bins=50, kde=True, color='blue')
plt.title('Amount Distribution for Non-Fraud (Class 0)')
plt.subplot(1,2,2)
sns.histplot(df[df['Class']==1]['Amount'], bins=50, kde=True, color='red')
plt.title('Amount Distribution for Fraud (Class 1)')
plt.tight_layout()
plt.show()
```

```python
# Calculate correlation matrix
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))
mask = np.triu(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            mask=mask, linewidths=0.5, annot_kws={"size": 7})
plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
print('Dataset shape: ' + str(df.shape))
print('Missing values in each column:')
print(df.isnull().sum())
```

```python
# Time analysis
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.scatter(df[df['Class']==0]['Time'], df[df['Class']==0]['Amount'], alpha=0.5, s=5, label='Non-Fraud')
plt.scatter(df[df['Class']==1]['Time'], df[df['Class']==1]['Amount'], alpha=0.5, s=5, label='Fraud')
plt.xlabel('Time (seconds)')
plt.ylabel('Amount')
plt.title('Transaction Amount vs Time')
plt.legend()

plt.subplot(1, 2, 2)
sns.histplot(data=df, x='Time', hue='Class', bins=50, alpha=0.7)
plt.title('Distribution of Transactions Over Time')
plt.tight_layout()
plt.show()
```

```python
# Feature importance analysis
fraud_mean = df[df['Class']==1].mean()
non_fraud_mean = df[df['Class']==0].mean()
mean_diff = pd.DataFrame({'Fraud': fraud_mean, 'Non-Fraud': non_fraud_mean})
mean_diff['Abs_Diff'] = abs(mean_diff['Fraud'] - mean_diff['Non-Fraud'])
mean_diff = mean_diff.sort_values('Abs_Diff', ascending=False)

# Plot top 10 discriminative features
plt.figure(figsize=(12, 6))
top_features = mean_diff.iloc[1:11] 
plt.bar(top_features.index, top_features['Abs_Diff'])
plt.xticks(rotation=90)
plt.title('Top 10 Discriminative Features (Absolute Mean Difference)')
plt.tight_layout()
plt.show()
```
